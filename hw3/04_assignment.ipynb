# Task 1

```python
# НАЧАЛО ВАШЕГО РЕШЕНИЯ
# Загружаем датасет для классификации
from datasets import load_dataset

raw_dataset = load_dataset("stanfordnlp/imdb")
print(raw_dataset)
print(set([raw_dataset["train"][i]["label"] for i in range(len(raw_dataset["train"]))]))
```

```python
# Загружаем токенизатор
from transformers import AutoTokenizer

model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

sample = raw_dataset["train"][0]["text"][:48]
encoded = tokenizer(sample, return_tensors="pt")

print(tokenizer.cls_token, tokenizer.sep_token, tokenizer.pad_token)
print(sample, "->", encoded["input_ids"])
```

```python
# Загружаем предобученный backbone
from transformers import AutoModel, AutoModelForSequenceClassification

base_encoder = AutoModel.from_pretrained(model_name)
clf_model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

print(base_encoder.__class__.__name__)
print(clf_model.__class__.__name__)
print(clf_model.num_labels)
```

```python
# Токенизируем датасет
def tokenize_batch(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=128,
    )

tokenized_dataset = raw_dataset.map(tokenize_batch, batched=True)
tokenized_dataset = tokenized_dataset.remove_columns(["text"])
tokenized_dataset = tokenized_dataset.rename_column("label", "labels")
tokenized_dataset.set_format("torch")

from torch.utils.data import DataLoader

dataloader = DataLoader(tokenized_dataset["train"], batch_size=24, shuffle=True)
```

```python
tokenized_dataset["train"]
```

```python
# Обучаем модель
import torch
import tqdm
from transformers import get_linear_schedule_with_warmup

device = "cuda" if torch.cuda.is_available() else "cpu"
clf_model = clf_model.to(device)

opt = torch.optim.AdamW(clf_model.parameters(), lr=2e-5, weight_decay=0.01)
EPOCH_NUM = 5
num_training_steps = EPOCH_NUM * len(dataloader)
num_warmup_steps = int(0.1 * num_training_steps)
scheduler = get_linear_schedule_with_warmup(opt, num_warmup_steps, num_training_steps)

losses = []

for ep in range(EPOCH_NUM):
    print("Epoch:", ep+1)
    cum_loss = 0.0
    for i, batch in enumerate(tqdm.tqdm(dataloader)):
        batch = {
            k: v.to(device)
            for k, v in batch.items()
            if k in ["input_ids", "attention_mask", "token_type_ids", "labels"]
        }
        opt.zero_grad()

        outputs = clf_model(**batch)
        loss = outputs.loss
        loss.backward()
        torch.nn.utils.clip_grad_norm_(clf_model.parameters(), 1.0)
        opt.step()
        scheduler.step()

        cum_loss += loss.item()

    losses.append(cum_loss / (i + 1))
    print("Loss:", losses[-1])
```

```python
# График loss
import matplotlib.pyplot as plt

plt.plot(losses)
plt.grid()
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Cross-Entropy graph")
```

```python
torch.save(clf_model.state_dict(),"./task1/task1_model.pt")
```

```python
# Считаем метрики
import evaluate

metric = evaluate.load("accuracy")

test_loader = DataLoader(tokenized_dataset["test"], batch_size=32, shuffle=False)
clf_model.eval()
clf_model.to(device)

for batch in test_loader:
    batch = {
        k: v.to(device)
        for k, v in batch.items()
        if k in ["input_ids", "attention_mask", "token_type_ids", "labels"]
    }
    with torch.no_grad():
        outputs = clf_model(**batch)
    preds = outputs.logits.argmax(dim=-1)
    metric.add_batch(predictions=preds, references=batch["labels"])

result = metric.compute()
print(result)
# КОНЕЦ ВАШЕГО РЕШЕНИЯ
```

```python
# НАЧАЛО ВАШЕГО РЕШЕНИЯ
import torch.nn as nn
from transformers import AutoConfig, AutoModel, TrainingArguments, Trainer
from transformers.modeling_outputs import SequenceClassifierOutput

class BertWithCustomHead(nn.Module):
    def __init__(
        self,
        config_name: str,
        num_labels: int,
        hidden: int = 768,
        dropout: float | None = 0.1,
    ):
        super().__init__()
        config = AutoConfig.from_pretrained(config_name, num_labels=num_labels)
        self.encoder = AutoModel.from_pretrained(config_name, config=config)
        self.dropout = nn.Dropout(dropout)

        self.head = nn.Sequential(
            nn.Linear(config.hidden_size, hidden),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(hidden, num_labels),
        )

        self.num_labels = num_labels

    def forward(
        self,
        input_ids=None,
        attention_mask=None,
        token_type_ids=None,
        labels=None,
    ):
        enc = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            return_dict=True,
        )
        cls = enc.last_hidden_state[:, 0, :]
        logits = self.head(self.dropout(cls))

        loss = None
        if labels is not None:
            loss = nn.functional.cross_entropy(logits, labels)

        return SequenceClassifierOutput(loss=loss, logits=logits)


model = BertWithCustomHead(model_name, num_labels=2)

args = TrainingArguments(
    output_dir="./task2",
    eval_strategy="no",  
    save_strategy="no",
    learning_rate=2e-5,
    weight_decay=0.01,
    num_train_epochs=5,
    per_device_train_batch_size=24,
    load_best_model_at_end=False,
    max_grad_norm=1.0,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_dataset["train"],
)
```

```python
trainer.train()
```

```python
torch.save(model.state_dict(), "task2/task2_model.pt")
```

```python
metric = evaluate.load("accuracy")
model.eval()
model.to(device)

for batch in test_loader:
    batch = {
        k: v.to(device)
        for k, v in batch.items()
        if k in ["input_ids", "attention_mask", "token_type_ids", "labels"]
    }
    with torch.no_grad():
        outputs = model(**batch)
    preds = outputs.logits.argmax(dim=-1)
    metric.add_batch(predictions=preds, references=batch["labels"])

result = metric.compute()
print(result)
# КОНЕЦ ВАШЕГО РЕШЕНИЯ
```

```python
# НАЧАЛО ВАШЕГО РЕШЕНИЯ
import numpy as np
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForTokenClassification,
    DataCollatorForTokenClassification,
    TrainingArguments,
    Trainer,
)
import evaluate
from functools import partial
from typing import List, Dict, Any

ds = load_dataset("conll2003")
label_list = ds["train"].features["ner_tags"].feature.names
num_labels = len(label_list)

def tokenize_and_align_labels(examples, tokenizer, label_all_tokens=False):
    tokenized = tokenizer(
        examples["tokens"],
        truncation=True,
        is_split_into_words=True,
    )
    aligned_labels = []
    for i, labels in enumerate(examples["ner_tags"]):
        word_ids = tokenized.word_ids(batch_index=i)
        prev_word_id = None
        example_labels = []
        for w_id in word_ids:
            if w_id is None:
                example_labels.append(-100) 
            elif w_id != prev_word_id:
                example_labels.append(labels[w_id])
            else:
                example_labels.append(labels[w_id] if label_all_tokens else -100)
            prev_word_id = w_id
        aligned_labels.append(example_labels)
    tokenized["labels"] = aligned_labels
    return tokenized

def prepare_dataset(model_name: str):
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
    fn = partial(tokenize_and_align_labels, tokenizer=tokenizer)
    dsm = ds.map(fn, batched=True, remove_columns=ds["train"].column_names)
    collator = DataCollatorForTokenClassification(tokenizer)
    return dsm, tokenizer, collator
```

```python
seqeval = evaluate.load("seqeval")

def compute_metrics(p, label_list: List[str]):
    logits, labels = p
    if isinstance(logits, tuple):
        logits = logits[0]
    preds = np.argmax(logits, axis=-1)

    true_predictions = []
    true_labels = []
    for pred, lab in zip(preds, labels):
        curr_preds = []
        curr_labels = []
        for p_i, l_i in zip(pred, lab):
            if l_i != -100:
                curr_preds.append(label_list[p_i])
                curr_labels.append(label_list[l_i])
        true_predictions.append(curr_preds)
        true_labels.append(curr_labels)

    results = seqeval.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

```python
def make_args(outdir: str):
    return TrainingArguments(
        output_dir=outdir,
        eval_strategy="steps",  
        save_strategy="no",
        learning_rate=2e-5,
        weight_decay=0.01,
        num_train_epochs=5,
        per_device_train_batch_size=24,
        load_best_model_at_end=False,
        max_grad_norm=1.0,
        logging_strategy="steps",
        logging_steps=50,
        eval_steps=250,
        report_to=[],                
    )

def train_and_eval(model_name: str, outdir: str) -> Dict[str, Any]:
    dsm, tokenizer, collator = prepare_dataset(model_name)
    print("Prepared data")
    model = AutoModelForTokenClassification.from_pretrained(
        model_name, num_labels=num_labels, id2label=dict(enumerate(label_list)),
        label2id={l:i for i, l in enumerate(label_list)}
    )
    args = make_args(outdir)
    print("Start training")
    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=dsm["train"],
        eval_dataset=dsm["test"],                  
        tokenizer=tokenizer,
        data_collator=collator,
        compute_metrics=lambda p: compute_metrics(p, label_list),
    )
    trainer.train()
    test_metrics = trainer.evaluate(eval_dataset=dsm["test"])

    return test_metrics, trainer
```

```python
bert_metrics, bert_trainer  = train_and_eval("bert-base-cased",  "./task3/ner_bert_cased")
bert_metrics
```

```python
xlnet_metrics, xlnet_trainer = train_and_eval("xlnet-base-cased", "./task3/ner_xlnet_cased")
xlnet_metrics
```

```python
import pandas as pd
import matplotlib.pyplot as plt

bert_hist = pd.DataFrame(bert_trainer.state.log_history)
xlnet_hist = pd.DataFrame(xlnet_trainer.state.log_history)

df_tr_bert = bert_hist.dropna(subset=["loss"])
df_tr_xlnet = xlnet_hist.dropna(subset=["loss"])

plt.figure()
plt.plot(df_tr_bert["step"], df_tr_bert["loss"], label="BERT", marker=".")
plt.plot(df_tr_xlnet["step"], df_tr_xlnet["loss"], label="XLNET", marker=".")
plt.xlabel("step"); plt.ylabel("train loss"); plt.title("Train loss"); plt.legend(); plt.grid(); plt.show()

df_ev_bert = bert_hist.dropna(subset=["eval_loss"])
df_ev_xlnet = xlnet_hist.dropna(subset=["eval_loss"])

for m in ["eval_loss","eval_precision","eval_recall","eval_f1","eval_accuracy"]:
    if m in df_ev_bert and m in df_ev_xlnet:
        plt.figure()
        plt.plot(df_ev_bert["step"], df_ev_bert[m], marker=".", label="BERT")
        plt.plot(df_ev_xlnet["step"], df_ev_xlnet[m], marker=".", label="XLNET")
        plt.xlabel("step"); plt.ylabel(m); plt.legend(); plt.grid(); plt.title(m.replace("_"," ").title())
        plt.show()
# КОНЕЦ ВАШЕГО РЕШЕНИЯ
```

# Выводы
Train loss: у BERT и XLNet почти совпадает и быстро падает. Обе модели хорошо оптимизируются.
Accuracy: различия минимальны (~0.982-0.985 у обеих).
F1: у XLNet слегка выше на поздних шагах.
Recall: XLNet стабильно выше.
Precision: близко; местами у BERT чуть выше, но в среднем равнозначно.

С точки зрения архитектуры:
1. Постановка задачи при обучении
    - XLNet учится предсказывать токены по разным пермутациям порядка, не вводя маскирование. За счет этого может быть меньше разрыв между пре-/дообучением, модель видит "естественные" контексты.
    - Для NER это помогает обрабатывать токены без маскированного шума и, следовательно, чаще находить сущности целиком. Тем самым повышается recall.

2. Relative positional encodings (Transformer-XL)
    - XLNet использует относительные позиции; они лучше обобщаются на разные длины и расстояния между токенами.
    - Относительное позиционное кодирование может дать большую устойчивость к обнаружению паттернов в тексте, что важно для задачи NER.
